```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["When were airplanes invented","when was the car invented","when was the train invented"]
table = pytrends.get_historical_interest(kw_list, year_start=2019, month_start=7, day_start=1, hour_start=0, year_end=2019, month_end=7, day_end=5, hour_end=18, cat=0, geo='', gprop='', sleep=0)
```

```python
table
```

```python
table.to_csv('trends.csv')
```
