```python
# libraries
#import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
 
# Get the data (csv file is hosted on the web)
gender_time = pd.read_csv(fname8)
```

```python
# And I need to transform my categorical column (continent) in a numerical value group1->1, group2->2...
gender_time['Subject']=pd.Categorical(gender_time['Subject'])
```

```python
# For each year:
for i in gender_time.Year.unique():
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
```

```python
# For each year:
for i in gender_time.Year.unique(): #Plots each year's data
    
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
    tmp=gender_time[gender_time.Year== i ]
    
    graphic = plt.scatter(tmp['Male Percent']*100, tmp['Total Taking Exams']/1000, c=tmp['Subject'].cat.codes, cmap="brg", alpha=0.6, edgecolors="white", linewidth=2)
    N=100
    
    # Add titles (main and on axis)
    plt.xlabel("Percentage of Students Taking Exam That Are Male")
    plt.ylabel("Total Student Taking AP Exam (Hundreds of Thousands) ")
    plt.title("Year: "+str(i) )
    plt.xlim(20,80)
    plt.ylim(0,600)
   
    import matplotlib.patches as mpatches
    
    #setting legend colors
    red_patch = mpatches.Patch(color='red', label='Humanities')
    blue_patch = mpatches.Patch(color='blue', label='Arts')
    green_patch = mpatches.Patch(color='limegreen', label='STEM')

    plt.legend(handles=[red_patch, blue_patch, green_patch])
    
    #saving each image for future use
    filename = "gender_time{0}.png".format(str(i))
    plt.savefig(filename, dpi=my_dpi)
```
